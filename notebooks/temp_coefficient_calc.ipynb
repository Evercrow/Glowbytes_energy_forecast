```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
big_df = pd.read_csv('../data/backup/train_dataset.csv')
```

```python
big_df = big_df[['date','temp','temp_pred']]
big_df.sample(30)
```

```python
cl_df = big_df.dropna()
print(f"Отсутствует {big_df.size-cl_df.size} значений")
```

```python
def calc_temp_delta_range(df:pd.DataFrame):
    t_fact = df['temp']#.to_numpy(copy=True)
    t_pred = df['temp_pred']#.to_numpy(copy=True)
    df['deltas'] = t_fact - t_pred
    delta_daily = df.groupby(df['date'])["deltas"].mean()
    print("Накопленная разница между температурами за день(ошибка)")
    print(delta_daily)
    print(delta_daily.describe())
    print("Выбрать коэффициент разбросa можно либо по медиане(часы уже учтены в шаге с группировкой по дню среднего значения),\n \
           либо по стандартному отклонению,с поправкой на час(/24)")
    coef = max(delta_daily.median(),delta_daily.std(ddof=1)/24)
    delta_range = (-coef,coef)
    return delta_range
calc_temp_delta_range(cl_df)
```
